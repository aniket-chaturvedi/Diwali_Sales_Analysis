```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #Data Visualizing
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace= True)
#To drop any value , axis=1 means vertical row
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
#.sum add all the null values if present
```

```python
df.shape
```

```python
df.dropna(inplace=True)
#drop null value 
```

```python
df.shape
```

```python
#chnge the data type
df['Amount']=df['Amount'].astype('int')
```

**Exploratory Data Analysis**

*Gender*

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)

for bars in ax.containers :
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index= False).sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Gender' , y='Amount', data = sales_gen,hue='Gender')
```

```python
df.columns
```

```python
ax = sns.countplot(data = df,x = 'Age Group',hue = 'Gender')

for bar in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs Age Group
sales_age = df.groupby(['Age Group'],as_index= False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data=sales_age,hue='Age Group')
```

*From above we can see that most of the buyers are of age group between 26-35 females*

**State**

```python
df.columns
```

```python
#total number of orders from the top 10 states

sales_states = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_states,hue='State')
```

```python
#total amount/sales of the top 10 state
sales_states = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_states,hue='State')
```

*From above graph we can see that most of the orders and total sales/amount are from Uttar Pradesh, Maharashtra, Karnataka and respectively*

**Marital Status**

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(4,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Marital_Status','Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)

sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_states,hue='Gender')
```

*From above graph we can see that most of the buyers are married(women) and they have high purchasing power*

**Occupation**

```python
df.columns
```

```python
sns.set(rc = { 'figure.figsize' : (20,5)})

ax = sns.countplot(data = df , x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Occupation'] ,as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.set(rc = { 'figure.figsize' : (20,5)})
sns.barplot(data = sales_states, x = 'Occupation', y = 'Amount',hue = 'Occupation')
```

*From above graph we can see that most of the buyers are working in IT, Healthcare, aviation sector respectively*

**Product Category**

```python
sns.set(rc = { 'figure.figsize' : (25,5)})

ax = sns.countplot(data = df , x = 'Product_Category',hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_states = df.groupby(['Product_Category'] ,as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False).head(10)

sns.set(rc = { 'figure.figsize' : (20,5)})
sns.barplot(data = sales_states, x = 'Product_Category', y = 'Amount',hue = 'Product_Category')
```

*From the above graph we can see that most of the sold product in Food, Clothing, Electronics and Gadgets category*

```python
df.columns
```

```python
sales_states = df.groupby(['Product_ID'] ,as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = { 'figure.figsize' : (20,5)})
sns.barplot(data = sales_states, x = 'Product_ID', y = 'Orders',hue = 'Product_ID')
```

**Conclusion**

*Married women age group 26-35 yrs from Uttar Pradesh, Maharashtra, Karnataka working in I, Healthcare & Aviation are more likely to buy products from Food, Clothing and Electronic*
